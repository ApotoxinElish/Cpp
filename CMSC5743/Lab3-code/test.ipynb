```python
d = np.load("pointcloud.npy")
d[0]
```

```python
dlist = d.tolist()
for i in range(64):
    for j in range(64):
        for k in range(64):
            dlist[i][j][k] = str(dlist[i][j][k])

output = []
for i in range(64):
    tmp = []
    for j in range(64):
        tmp.append(",".join(dlist[i][j]))
    output.append(",".join(tmp))
with open("pointcloud.csv", "w") as f:
    f.write("\n".join(output))
```
